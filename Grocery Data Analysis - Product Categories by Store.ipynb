```python
# Create a table to show the count of certain categories by store name
counts = file.groupby(['STORE_NAME', 'CATEGORY']).size().unstack(fill_value=0)
```

```python
counts
```

```python
import matplotlib.pyplot as plt
```

```python
# Get the store names and categories from the DataFrame
stores = counts.index.tolist()
categories = counts.columns.tolist()
```

```python
print(len(categories))
```

```python
import random

# Generate random colors
colors = [random.choice(range(256**3)) for _ in range(len(categories))]  

# Convert the random colors to hex format
colors = ['#' + format(color, '06x') for color in colors]
```

```python
# Calculate the total count for each store
store_totals = counts.sum(axis=1)

# Calculate the number of rows and columns for the subplot grid
num_rows = int(len(stores) ** 0.5)
num_cols = int(len(stores) / num_rows) + (len(stores) % num_rows > 0)

# Set the figure size
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 10))  # Adjust the width and height as needed

# Create a legend for categories outside the image
legend_labels = []
legend_colors = []

# Iterate over each store
for i, store in enumerate(stores):
    # Calculate the percentage for each category
    percentages = counts.loc[store] / store_totals.loc[store] * 100

    # Get the corresponding subplot axis
    if num_rows > 1 and num_cols > 1:
        ax = axes[i // num_cols, i % num_cols]
    elif num_rows > 1:
        ax = axes[i]
    elif num_cols > 1:
        ax = axes[i]
    else:
        ax = axes

    # Create a pie chart for the current store
    wedges, labels, _ = ax.pie(percentages, colors=colors, autopct=lambda pct: f'{pct:.1f}%' if pct > 10 else None,
                               startangle=90)
    
    # Collect the labels and colors for the legend
    legend_labels.extend(labels)
    legend_colors.extend(colors)
    
    # Set the font size for label text
    ax.set_title(store, fontsize=10)
    ax.tick_params(labelsize=8)

    # Ensure equal aspect ratio to make pie chart circular
    ax.axis('equal')

# Remove any empty subplots
if len(stores) < num_rows * num_cols:
    if num_rows > 1 and num_cols > 1:
        for j in range(len(stores), num_rows * num_cols):
            fig.delaxes(axes[j // num_cols, j % num_cols])
    else:
        for j in range(len(stores), num_rows + num_cols):
            fig.delaxes(axes[j])

# Create a legend outside the image for the categories
fig.legend(legend_labels, labels=categories, loc='lower center', ncol=2, frameon=False, bbox_to_anchor=(0.5, -0.15))

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate the percentage for each category for each store
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Initialize the bottom array
bottom = np.zeros(len(stores))

# Iterate over each category
for i, category in enumerate(categories):
    # Get the category percentages for all stores
    category_percentages = percentages[category].values

    # Plot the bar for the category
    p = ax.bar(stores, category_percentages, width=0.8, bottom=bottom, color=colors[i], label=category)

    # Update the bottom array for the next category
    bottom += category_percentages

# Set the y-axis limits
ax.set_ylim(0, 100)

# Set the y-axis label
ax.set_ylabel('Percentage (%)')

# Set the title
ax.set_title('Category Distribution by Store')

# Create a legend outside the graph
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Adjust x-axis tick label font size and rotation
ax.tick_params(axis='x', labelsize=8, rotation=45)

# Adjust x-axis label positioning
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
